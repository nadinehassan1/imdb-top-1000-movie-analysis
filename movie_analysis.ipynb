## Task 2 : Data preparation ##

```python
df['Gross'] = df['Gross'].str.replace(',', '')
df['Gross'] = pd.to_numeric(df['Gross'])
```

##### Handling Missing Values

The dataset contains missing values in three columns: Certificate, Meta_score, and Gross.

- Certificate is a categorical variable and missing values were replaced with "Unknown".
- Meta_score is a numerical variable, and missing values were filled using the median value.
- Gross revenue was first cleaned by removing commas and converting the column to numeric format, then missing values were filled using the median.

This ensures that the dataset contains no missing values before proceeding to further analysis.

```python
df['Certificate']= df['Certificate'].fillna('Unknown')
df['Meta_score'] = df['Meta_score'].fillna(df['Meta_score'].median())
df['Gross'] = df['Gross'].fillna(df['Gross'].median())
```

## Runtime Data Cleaning

The Runtime column contained values in the format "142 min".  
To enable numerical analysis and correlation calculations, the text "min" was removed and the column was converted to an integer data type representing runtime in minutes.

```python
df['Runtime'] = df['Runtime'].str.replace(' min','')
df['Runtime'] = df['Runtime'].astype(int)
```

```python
Q1 = df['No_of_Votes'].quantile(0.25)
Q3 = df['No_of_Votes'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['No_of_Votes'] < lower) | (df['No_of_Votes'] > upper)]

outliers
```

## Feature Engineering

Feature engineering was applied to create additional insights from the data.

For example, the number of votes was transformed into popularity categories using binning. This allowed movies to be grouped into **Low Popularity**, **Medium Popularity**, and **High Popularity** groups, making it easier to analyze how popularity relates to ratings.

## Binning

Binning was applied to the number of votes feature to categorize movies based on their popularity. Continuous vote counts were grouped into categories representing low, medium, and high popularity levels. This transformation helps simplify the analysis and makes it easier to explore relationships between movie popularity and ratings.

```python
bins = [0,100000,500000,2500000]
labels = ['Low Popularity','Medium Popularity','High Popularity']

df['Vote_Category'] = pd.cut(df['No_of_Votes'], bins=bins, labels=labels)
```

```python
df['Vote_Category'].value_counts()
```

## Model Selection

A **Linear Regression model** was chosen to predict IMDb ratings based on several numerical features:

- Meta Score
- Number of Votes
- Runtime

Linear regression is suitable for this problem because the target variable (IMDb rating) is continuous, and the goal is to understand how these features influence the predicted rating.

```python
X = df[['Meta_score','No_of_Votes','Runtime']]
y = df['IMDB_Rating']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

train model

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

Evaluation

```python
print("R² Score:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
```

The R² score of **0.33** indicates that about **33% of the variation in IMDb ratings** can be explained by the selected features. This suggests that these factors have some influence on movie ratings, but other variables not included in the model also play a significant role.

The **MAE of 0.18** means that, on average, the model’s predictions differ from the actual IMDb ratings by approximately **0.18 rating points**, indicating relatively small prediction errors.

## Model Limitations

While the model captures some relationships between the selected features and IMDb ratings, the relatively low R² score indicates that additional variables may influence movie ratings.

Factors such as:

- Genre
- Director reputation
- Cast popularity
- Release year
- Production budget

could further improve the predictive performance of the model.

### Actual vs Predicted IMDb Ratings

This plot compares the **actual IMDb ratings** with the **ratings predicted by the linear regression model**.

The red diagonal line represents **perfect predictions**, where the predicted value equals the actual value. Points closer to this line indicate more accurate predictions.

Most data points lie relatively close to the diagonal line, suggesting that the model is able to capture some of the relationship between the selected features and movie ratings. However, the visible spread around the line indicates that prediction errors still exist and other factors may influence IMDb ratings.

```python
plt.figure(figsize=(8,5))

sns.scatterplot(x=y_test, y=y_pred)
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()],
         color='red')
plt.xlabel("Actual IMDb Rating")
plt.ylabel("Predicted IMDb Rating")
plt.title("Actual vs Predicted IMDb Ratings")
```

### Residual Analysis

This residual plot shows the difference between the **actual IMDb ratings** and the **ratings predicted by the model**.

Residuals represent the prediction error (Actual Rating − Predicted Rating). The horizontal red line represents **zero error**.

Ideally, residuals should be randomly distributed around zero, indicating that the model does not consistently overestimate or underestimate the ratings.

In this case, the residuals are scattered around the zero line, suggesting that the model captures some patterns in the data. However, the spread of points indicates that prediction errors still exist and that additional features may improve the model's performance.

```python
residuals = y_test - y_pred

plt.figure(figsize=(8,5))
sns.scatterplot(x=y_pred, y=residuals)

plt.axhline(0, color='red')

plt.xlabel("Predicted IMDb Rating")
plt.ylabel("Residuals")
plt.title("Residual Plot")
```

Most residuals fall roughly between -0.4 and +0.6
That aligns well with your MAE ≈ 0.18, meaning prediction errors are relatively small.
So your model performance is reasonable for this dataset.
